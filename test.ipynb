```python
# List containing escape sequences
escape_list = ["\n", "\t", "\\", "\'", "\""]

# Explanation of escape sequences:
# \n - Newline
# \t - Tab
# \\ - Backslash
# \' - Single quote
# \" - Double quote

# Printing the list
for escape in escape_list:
    print(f"Escape sequence: {repr(escape)} -> Output: {escape}")
```

```python

# A list inside a list (nested list)
nested_list = [1, 2, [3, 4, 5], 6, [7, 8]]

# Accessing elements
print(nested_list[2])       # Output: [3, 4, 5] (the inner list)
print(nested_list[2][1])    # Output: 4 (element at index 1 of the inner list)
print(nested_list[4][0])    # Output: 7 (element at index 0 of the second inner list)

# Iterating through the nested list
for item in nested_list:
    print(item)
```
